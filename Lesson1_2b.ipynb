```python
import requests
url = "https://data.ntpc.gov.tw/api/datasets/010E5B15-3823-4B20-B401-B1CF000550C5/json?page=0&size=10000"
respones = requests.get(url)
dataList = respones.json()
print(dataList)
```

```python
print(len(dataList))
for item in dataList:
    print("站點：{}\n可借：{}\n可還：{}\n更新時間：{}\n---------------------------------------".format(item['sna'],item['sbi'],item['bemp'],item['mday']))
```

```python
import csv
with open("output2.csv","w",newline="",encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["站點","可借","可還","更新時間"])
    for item in dataList:
        writer.writerow([item['sna'],item['sbi'],item['bemp'],item['mday']])
print("finish")
```
